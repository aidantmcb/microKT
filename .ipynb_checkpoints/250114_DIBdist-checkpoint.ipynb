```python
import numpy as np 
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.table import Table 
 
from astropy.stats import sigma_clip, sigma_clipped_stats
```

```python
fname = '../240723_EmissionMaps/CA_meta_LSR_uncerts.fits'
tab = Table(fits.open(fname)[1].data)
```

```python
print(tab.colnames)
```

```python
fig, ax = plt.subplots()
hist = np.histogram2d(tab['DIST'], tab['DIB_EQW'], range  = [[350, 650], [0, 0.3]], bins=(25, 25))
ax.imshow(hist[0].T, origin = 'lower', extent = [350, 650, 0, 0.3], cmap = 'bone_r')
ax.set_aspect(500)
ax.set_xlim(350, 650)
ax.set_ylim(0, 0.3)
ax.set_xlabel('Distance (pc)')
ax.set_ylabel(r'DIB EW ($\AA$)')
plt.show()
```

```python
line = lambda x, m: m*x + 0
from scipy.optimize import curve_fit
popt, pcov = curve_fit(line, tab['DIST'], tab['DIB_EQW'], nan_policy = 'omit')

fig, ax = plt.subplots()
ax.imshow(hist[0].T, origin = 'lower', extent = [350, 650, 0, 0.3], cmap = 'bone_r')
ax.set_aspect(900)
ax.plot(np.linspace(350, 650, 100), line(np.linspace(350, 650, 100), *popt), color = 'red')
ax.set_xlim(350, 650)
ax.set_ylim(0, 0.3)
ax.set_xlabel('Distance (pc)')
ax.set_ylabel(r'DIB EW ($\AA$)')

sigclip = sigma_clip(tab['DIB_EQW'] - line(tab['DIST'], *popt), sigma = 1.5, maxiters = 5)
ax.scatter(tab['DIST'][sigclip.mask], tab['DIB_EQW'][sigclip.mask], color = 'red', s = 1)

ax.scatter(tab['DIST'][~sigclip.mask], tab['DIB_EQW'][~sigclip.mask], color = 'blue', s = 1)
sigclip
```
