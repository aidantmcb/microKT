```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys 
import glob

from astropy.table import Table
```

```python
data = []
names = []
for fname in glob.glob('/uufs/astro.utah.edu/common/home/u1371365/nanoKT_v1/spearmanINJECTION*'):
    data.append(np.loadtxt(fname))
    names.append(fname.split('T_')[-1].split('.csv')[0])
```

```python
labels = []
for name in names:
    for item in ['_stat', '_p']:
        labels.append(name + item)
tab = Table(np.hstack(data), names = labels)
```

```python
tab.write('spearman_outputs.csv', format ='csv', overwrite=True)
```

```python
fig, ax = plt.subplots()

for i in range(len(names), ):
    stat_lab = labels[2 * i]
    p_lab = labels[2 * i+1]
    ax.scatter(tab[stat_lab], tab[p_lab], marker = 'o', label = names[i])
    ax.set_xlabel("SPEARMAN STAT")
    ax.set_ylabel("SPEARMAN p")
    ax.legend()
plt.show()
```
